*18º Interação - Crie um script para identificar as colunas com outliers. Mostre o resultado da análise de cada coluna.*



---



Código gerado pela 18º interação:

Observação: Nesse código gerado, ele faz anaálise domente com a técnica do desvio padrão. Caso queria, pode solicitar para elaborar um código com mais ténicas de detecção.

Lembrando também que é somente para colunas com dados numéricos.

```python
# Liste as colunas numéricas para análise de outliers
numeric_columns = ['age', 'final-weight', 'education-num', 'capital-gain', 'capital-loos', 'hour-per-week']

# Defina um limite para identificar outliers (por exemplo, 3 desvios padrão)
outlier_threshold = 3

# Função para identificar outliers em uma coluna
def identify_outliers(column):
    mean = column.mean()
    std = column.std()
    lower_limit = mean - outlier_threshold * std
    upper_limit = mean + outlier_threshold * std
    outliers = (column < lower_limit) | (column > upper_limit)
    return outliers

# Identifique e analise outliers em cada coluna numérica
for column_name in numeric_columns:
    column = data[column_name]
    outliers = identify_outliers(column)

    # Exiba o resultado da análise da coluna
    print(f"Análise da coluna '{column_name}':")
    print(f" - Número total de outliers: {outliers.sum()}")
    print(f" - Valor médio: {column.mean()}")
    print(f" - Desvio padrão: {column.std()}")
    print(f" - Limite inferior para outliers: {column.mean() - outlier_threshold * column.std()}")
    print(f" - Limite superior para outliers: {column.mean() + outlier_threshold * column.std()}")
    print("\n")

# Exiba um resumo das colunas com outliers
print("Colunas com outliers:")
for column_name in numeric_columns:
    column = data[column_name]
    if identify_outliers(column).any():
        print(f"'{column_name}' tem outliers.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Defina o tamanho padrão das figuras
plt.rcParams['figure.figsize'] = (12, 6)

# 1. Gráfico de Barras (Bar Chart)
plt.subplot(3, 3, 1)
sns.countplot(x='income', data=data)
plt.title('Gráfico de Barras: Renda')

# 2. Gráfico de Barras Empilhadas (Stacked Bar Chart)
plt.subplot(3, 3, 2)
sns.countplot(x='income', hue='sex', data=data)
plt.title('Gráfico de Barras Empilhadas: Renda por Sexo')

# 3. Gráfico de Pizza (Pie Chart)
plt.subplot(3, 3, 3)
income_counts = data['income'].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gráfico de Pizza: Distribuição de Renda')

# 4. Gráfico de Rosca (Donut Chart)
plt.subplot(3, 3, 4)
income_counts = data['income'].value_counts()
plt.pie(income_counts, labels=income_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Gráfico de Rosca: Distribuição de Renda')

# 5. Gráfico de Barras Horizontais (Horizontal Bar Chart)
plt.subplot(3, 3, 5)
sns.countplot(y='income', data=data)
plt.title('Gráfico de Barras Horizontais: Renda')

# 6. Gráfico de Setores (Treemap)
import squarify
plt.subplot(3, 3, 6)
income_counts = data['income'].value_counts()
squarify.plot(sizes=income_counts, label=income_counts.index, alpha=0.7)
plt.axis('off')
plt.title('Gráfico de Setores: Distribuição de Renda')

# 7. Gráfico de Pareto
plt.subplot(3, 3, 7)
income_counts = data['income'].value_counts()
income_cumulative = income_counts.cumsum() / income_counts.sum()
sns.barplot(x=income_counts.index, y=income_cumulative, color='skyblue')
plt.title('Gráfico de Pareto: Distribuição de Renda')

# 8. Gráfico de Frequência (Frequency Plot)
plt.subplot(3, 3, 8)
sns.histplot(data=data, x='income', bins=2, discrete=True)
plt.title('Gráfico de Frequência: Renda')

# 9. Gráfico de Caixas (Box Plot)
plt.subplot(3, 3, 9)
sns.boxplot(x='income', y='age', data=data)
plt.title('Gráfico de Caixas: Idade por Renda')

# Ajuste o layout das figuras
plt.tight_layout()

# Exiba os gráficos
plt.show()
```

Código gerado pela 48º interação:

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis, LinearDiscriminantAnalysis
from sklearn.cluster import KMeans
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.neighbors import NearestCentroid

# Defina uma lista de algoritmos
algorithms = [
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("K-Nearest Neighbors", KNeighborsClassifier(n_neighbors=5)),
    ("Support Vector Machine", SVC(kernel="linear")),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Random Forest", RandomForestClassifier(n_estimators=100)),
    ("Gradient Boosting", GradientBoostingClassifier()),
    ("AdaBoost", AdaBoostClassifier()),
    ("Naive Bayes", GaussianNB()),
    ("Neural Network", MLPClassifier(max_iter=1000)),
    ("Quadratic Discriminant Analysis", QuadraticDiscriminantAnalysis())
]

# Crie dicionários para armazenar as métricas
accuracy_scores = {}
precision_scores = {}
recall_scores = {}
f1_scores = {}

# Itere sobre os algoritmos
for name, classifier in algorithms:
    # Treine o modelo
    classifier.fit(X_train, y_train)

    # Faça previsões no conjunto de teste
    y_pred = classifier.predict(X_test)

    # Calcule as métricas
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Armazene as métricas nos dicionários
    accuracy_scores[name] = accuracy
    precision_scores[name] = precision
    recall_scores[name] = recall
    f1_scores[name] = f1

# Apresente um resumo das métricas
print("Resumo das Métricas para os Algoritmos:")
for name, _ in algorithms:
    print(f"Algoritmo: {name}")
    print(f"Acurácia: {accuracy_scores[name]:.2f}")
    print(f"Precisão: {precision_scores[name]:.2f}")
    print(f"Recall: {recall_scores[name]:.2f}")
    print(f"F1-Score: {f1_scores[name]:.2f}")
    print()
```

###Validação cruzada

*50º Interação - O que é a validação cruzada?*



---



*51º Interação - Agora eu preciso de um script para fazer a validação cruzada, porém, utilizando somente os algoritmos que tiverem os 4 melhores resultados. Mostre no final um resumo com os resultados de cada algoritmo. Inclua o desvio padrão no script anterior, também o mostrando no final para cada algoritmo.*

**Observação: Quanto menor o desvio padrão, mais consistente é o algoritmo**



---



Código gerado pela 51º interação:

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Defina os melhores algoritmos com base nos resultados anteriores
melhores_algoritmos = {
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "Neural Network": MLPClassifier(max_iter=1000)
}

# Dicionário para armazenar os resultados da validação cruzada
resultados = {}

# Realize a validação cruzada para cada um dos melhores algoritmos
for nome, algoritmo in melhores_algoritmos.items():
    # Realize a validação cruzada com 5 folds
    scores = cross_val_score(algoritmo, X_train, y_train, cv=5, scoring='accuracy')

    # Calcule as métricas médias
    acuracia_media = scores.mean()

    # Calcule o desvio padrão das métricas
    desvio_padrao_acuracia = scores.std()

    # Treine o modelo no conjunto de treinamento completo
    algoritmo.fit(X_train, y_train)

    # Faça previsões no conjunto de teste
    previsoes = algoritmo.predict(X_test)

    # Calcule outras métricas
    precisao = precision_score(y_test, previsoes)
    recall = recall_score(y_test, previsoes)
    f1 = f1_score(y_test, previsoes)

    # Armazene os resultados no dicionário
    resultados[nome] = {
        "Acurácia Média": acuracia_media,
        "Desvio Padrão da Acurácia": desvio_padrao_acuracia,
        "Precisão": precisao,
        "Recall": recall,
        "F1-Score": f1
    }

# Imprima um resumo dos resultados
print("Resumo das Métricas para os Melhores Algoritmos:")
for nome, metricas in resultados.items():
    print(f"Algoritmo: {nome}")
    print(f"Acurácia Média: {metricas['Acurácia Média']:.2f}")
    print(f"Desvio Padrão da Acurácia: {metricas['Desvio Padrão da Acurácia']:.5f}")
    print(f"Precisão: {metricas['Precisão']:.2f}")
    print(f"Recall: {metricas['Recall']:.2f}")
    print(f"F1-Score: {metricas['F1-Score']:.2f}")
    print()
```
